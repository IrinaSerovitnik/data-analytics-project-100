Подключение необходимых библиотек

```python
import pandas as pd
import requests
```

Запрос данных по визитам по API за период с 2023-03-01 по 2023-09-01

```python
visits = requests.get('https://data-charts-api.hexlet.app/visits?begin=2023-03-01&end=2023-09-01')
```

Создание датафрейма по загруженным данным по визитам

```python
visits_data = pd.DataFrame(visits.json())
```

Просмотр среза данных по визитам

```python
visits_data.head()
```
